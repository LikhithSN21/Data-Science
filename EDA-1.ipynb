##we have the Approach to treat the missing values 
1. either we delete the NAN values in the columns.
2.take the mean value of the particular column and fill it with the mean value


##in this dataset since all the columns are interdependent in the one or more way i would like to add the mean values to the missing columns

```python
LBmean=df_cleaned1['LB'].mean()
```

```python
LBmean
```

```python
df_cleaned1['LB']=df_cleaned1['LB'].fillna(LBmean)
```

```python
ACmean=df_cleaned1['AC'].mean()
```

```python
df_cleaned1['AC']=df_cleaned1['AC'].fillna(ACmean)
```

```python
DSmean=df_cleaned1['DS'].mean()
```

```python
df_cleaned1['DS']=df_cleaned1['DS'].fillna(DSmean)
```

```python
DPmean=df_cleaned1['DP'].mean()
```

```python
df_cleaned1['DP']=df_cleaned1['DP'].fillna(DPmean)
```

```python
MLTVmean=df_cleaned1['MLTV'].mean()
```

```python
df_cleaned1['MLTV']=df_cleaned1['MLTV'].fillna(MLTVmean)
```

```python
widthmean=df_cleaned1['Width'].mean()
```

```python
df_cleaned1['Width']=df_cleaned1['Width'].fillna(widthmean)
```

```python
Tmean=df_cleaned1['Tendency'].mean()
```

```python
df_cleaned1['Tendency']=df_cleaned1['Tendency'].fillna(Tmean)
```

```python
NSPmean=df_cleaned1['NSP'].mean()
```

```python
df_cleaned1['NSP']=df_cleaned1['NSP'].fillna(NSPmean)
```
